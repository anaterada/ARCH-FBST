## Modelo ARCH(1)

Assumindo os priors $\alpha_0 \sim Exp(1)$ e $\alpha_1 \sim Exp(1)$.

```python
def get_posterior_density(retornos, theta_hypothesis=[None, None]):
    
    def log_posterior_density(theta):
        a0, a1 = theta[0], theta[1]
        
        if not theta_hypothesis[0] is None:
            a0 = theta_hypothesis[0]
        if not theta_hypothesis[1] is None:
            a1 = theta_hypothesis[1]  
        
        L = 0
        L += -1 # Priori - Alpha_0
        L += -1 # Priori - Alpha_1
        L += -(len(retornos)-1)/2*np.log(2*np.pi)

        for t in range(1, len(retornos)):
            L += -1/2*np.log(a0 + a1*(retornos[t-1]**2))
            L += -1/2*retornos[t]**2/(a0 + a1*(retornos[t-1]**2))
            
        return L
    
    return log_posterior_density
```

```python
posterior = get_posterior_density(retornos)
```

```python
sampler = MetropolisSampler(posterior, starting_point=[0.5, 0.5], kernel_sigma=[.015, .015], 
                            restrictions=[[.01, np.inf], [.01, np.inf]])
sampler.warm(500)
samples = sampler.sample(10_000, show_progress=True)
```

```python
plt.figure(figsize=(18, 5))
plt.subplot(1, 2, 1)

# Histograma 2d
plt.hist2d(samples[:,0], samples[:,1], bins = [40, 40])[1][0]
plt.title('Amostras da distribuição posteriori')
plt.xlabel('$\\alpha_0$', fontsize = 16); plt.ylabel('$\\alpha_1$', fontsize = 16)

# Trace plot
plt.subplot(1, 2, 2)
plt.plot(samples[:, 1])
plt.plot(samples[:, 0])
plt.title('Trajetória da cadeia')
```

## FBST [https://pt.wikipedia.org/wiki/FBST]

Vamos testar a hipótese de estacionariedade forte da volatilidade usando o FBST

$H_0: \alpha_1 = 0$

$H_1: \alpha_1 \neq 0$

```python
from src.inference import evidence
from scipy.optimize import minimize
```

Verossimilhança restrita

```python
restricted_loglik = get_posterior_density(retornos, theta_hypothesis=[None, 0])
neg_restricted_loglike = lambda theta_0: -1*restricted_loglik([theta_0, None])
```

Maximização da verossimilhança restrita à $H_0$

```python
opt_min = minimize(neg_restricted_loglike, 0.2)
restricted_maximum_loglikelihood_value = -1*opt_min.fun
```

Cálculo da evidência

```python
def evidence(simulations, unrestricted_loglikelihood_fn, restricted_maximum_loglikelihood_value):
    '''
        Calcula a medida de evidência com as simulações geradas, a função objetivo irrestrita e 
        o valor máximo da função objetivo restrita a hipótese do teste.
    '''
    unrestricted_loglikelihood_value = np.array([unrestricted_loglikelihood_fn(sample) for sample in simulations])
    return (1 - np.average(unrestricted_loglikelihood_value >= restricted_maximum_loglikelihood_value))
```

```python
evidence(samples, posterior, restricted_maximum_loglikelihood_value)
```
